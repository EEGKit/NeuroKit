```python
import neurokit2 as nk

sampling_rate=100
rsp = nk.rsp_simulate(duration=30, sampling_rate=sampling_rate, method="breathmetrics")
rsp_cleaned = nk.rsp_clean(rsp, sampling_rate=sampling_rate)
rsp_epochs = nk.rsp_segment(rsp_cleaned, sampling_rate=sampling_rate, show=True)
```

```python
peak_signal, info = nk.rsp_peaks(rsp_cleaned, sampling_rate=sampling_rate)
peaks = info["RSP_Peaks"]
breaths, average_rr = nk.signal_cyclesegment(rsp_cleaned, peaks, sampling_rate=sampling_rate)
```
